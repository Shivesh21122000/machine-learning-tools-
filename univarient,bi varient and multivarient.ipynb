```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df=sns.load_dataset('iris')
```

```python
df
```

```python
sns.pairplot(df,hue='species')
```

```python
df_setosa=df.loc[df['species']=='setosa']
```

```python
df_setosa.head()
```

```python
df_virginica=df[df['species']=='virginica']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'*')
```

```python
sns.FacetGrid(df,hue="species",size=5).map(plt.scatter,'sepal_length','sepal_width').add_legend();
plt.show()
sns.FacetGrid(df,hue="species",size=5).map(plt.scatter,"petal_length","sepal_width").add_legend();
plt.show()
```

```python
sns.FacetGrid(df,hue="species",size=6).map(plt.scatter,'petal_length','petal_width').add_legend();
plt.show()
```

```python
plt.plot(df_setosa['petal_length'],np.zeros_like(df_setosa['petal_length']),'*')
```

```python
sns.FacetGrid(df,hue='species',size=5).map(plt.scatter,'sepal_length','petal_length').add_legend();
plt.show()
```
